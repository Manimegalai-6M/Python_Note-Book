```python
import calendar

year = 2023
month = 5

# Create calendar object
cal = calendar.monthcalendar(year, month)

print(f"Calendar for {month}/{year} (without Sundays):\n")

# Print week by week (ignoring Sunday column which is index 6)
for week in cal:
    for day in week[:-1]:   # exclude last column (Sunday)
        if day == 0:
            print("   ", end=" ")
        else:
            print(f"{day:2d}", end=" ")
    print()
```

```python
import calendar

# Input
year = 2023
month = 5

# ANSI escape codes for red color
RED = "\033[91m"
RESET = "\033[0m"

# Generate month calendar
cal = calendar.monthcalendar(year, month)

print(f"Calendar for {month}/{year} (Sundays in Red):\n")
print("Mo Tu We Th Fr Sa Su")  # Header

for week in cal:
    for i, day in enumerate(week):
        if day == 0:   # Empty days
            print("  ", end="  ")
        elif i == 6:   # Sunday (last column)
            print(f"{RED}{day:2d}{RESET}", end="  ")
        else:
            print(f"{day:2d}", end="  ")
    print()
```
