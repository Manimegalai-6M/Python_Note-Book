```python
#find the whether a year is leap year or not 
```

```python
n=int(input("n:"))
if n%4==0:
    if n%100==0:
        if n%400==0:
            print("leap year")
        else:
            print("not leap year")
    else:
        print("leap year")
else:
    print("not leap year")
```

```python
n=int(input("n:"))
m=int(input("m:"))
for i in range(n,m+1):
    p=len(str(i))
    if i==sum(int(d)**p for d in str(i)):
        print(i)
```

```python
n=int(input("n:"))
m=int(input("m:"))
for i in range(n,m+1):
    power=len(str(i))
    value=0
    temp=i
    while temp>0:
        digit=temp%10
        value=value+digit**power
        temp=temp//10
    if i==value:
        print(i)
```
