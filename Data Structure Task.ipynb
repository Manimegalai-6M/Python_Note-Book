```python
'''47.Search an element using binary search (recursive method).'''

def binary_search(arr, low, high, key):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == key:
        return mid
    elif key < arr[mid]:
        return binary_search(arr, low, mid - 1, key)
    else:
        return binary_search(arr, mid + 1, high, key)

arr = [10, 20, 30, 40, 50]
print(binary_search(arr, 0, len(arr) - 1, 30))
```

```python
'''66.Create a linked list and count the number of nodes in it.'''

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def add(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def count_nodes(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count

# Testing
ll = LinkedList()
ll.add(10)
ll.add(20)
ll.add(30)

print("Number of nodes:", ll.count_nodes())

```

```python
'''68.Create a linked list and print only the first and last node values.'''

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def add(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                return "Value found"
            temp = temp.next
        return "Value not found"


# Testing
ll = LinkedList()
ll.add(5)
ll.add(15)
ll.add(25)

print(ll.search(15))  # Output: Value found
print(ll.search(100)) # Output: Value not found
```

```python
'''70.Create a linked list and display the data of each node along with its next reference.'''

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def add(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def display_with_next(self):
        temp = self.head
        while temp:
            print("Data:", temp.data, "| Next:", temp.next)
            temp = temp.next

# Testing
ll = LinkedList()
ll.add(1)
ll.add(2)
ll.add(3)

ll.display_with_next()
```
