```python
#How to access the dictionary key and value
```

```python
a={"ame":"mani","age":12}
for i in a.keys():
    print(i)
for j in a.values():
    print(j)
```

```python
a={"ame":"mani","age":12}
b={"dept":"aids","year":4}
for i,k in a.keys(),b.keys():
    print(i,k,sep="\n")
for j,l in a.values(),b.values():
    print(j,l,sep="\n")
```

```python
a={"ame":"mani","age":12}
for i in a.items():
    print(i)
```

```python
a={"ame":"mani","age":12}
b={"dept":"aids","year":4}
for i,j in a.items(),b.items():
    print(i,j,sep="\n")
```

```python
a=[[2,3],[5,6,7],[8,9,0]]
b=[]
for i in a:
    for j in i:
        b.append(j)
print(b)
        
```

```python
#print pyramid pattern
```

```python
rows=int(input("n:"))
for i in range(1,rows+1):
    for j in range(1,i+1):
        print("*",end="")
    print("\n")
    
```
