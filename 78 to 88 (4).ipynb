```python
#ask 3: Count digits using recursion
def count_digits(n):
    n = abs(n)
    if n < 10:
        return 1
    return 1 + count_digits(n // 10)

num = 123456
print(f"{num} has {count_digits(num)} digits")
```

```python
num = 246813
even, odd = 0, 0

for digit in str(abs(num)):
    if int(digit) % 2 == 0:
        even += 1
    else:
        odd += 1

print(f"Even digits = {even}, Odd digits = {odd}")
```

```python
#Task 6: Count how many times each digit appears
num = 1122334455
freq = {}

for digit in str(num):
    freq[digit] = freq.get(digit, 0) + 1

for d, c in freq.items():
    print(f"{d} → {c} times")
```
