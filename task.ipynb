# Tasks Using abs() in Python

```python
'''1.Task 1: Find the absolute value of a number
Input: num = -25
Output: 25'''

num=-25
print(abs(num))
```

```python
'''2.Task 2: Calculate the distance between two numbers on a number line
Input: x = -7, y = 5
Output: Distance = 12'''

x,y=-7,5
print("Distance =",abs(x-y))
```

```python
'''3.Task 3: Sum of absolute differences in a list of numbers
Input: [3, -7, 4, -2]
Output: Sum of absolute values = 16'''

a=[3, -7, 4, -2]
d=sum(abs(i) for i in a)
print('Sum of absolute values =',d) 
```

```python
'''3.Task 3: Sum of absolute differences in a list of numbers
Input: [3, -7, 4, -2]
Output: Sum of absolute values = 16'''

a=[3, -7, 4, -2]
s=[]
for i in a:
    b=abs(i)
    s.append(b)
print('Sum of absolute values =',sum(s))
```

```python
'''4.Task 4: Find the closest number to zero in a list
Input: [-10, 3, -2, 7, -5]
Output: -2'''

a=[-10, 3, -2, 7, -5]
print(min(a,key=abs))
```

```python
'''5.Task 5: Sort a list of numbers by their absolute values
Input: [-10, 3, -2, 7, -5]
Output: [-2, 3, -5, 7, -10]'''

a=[-10, 3, -2, 7, -5]
print(sorted(a,key=abs))
```

```python
'''6.Task 6: Check if two numbers are the same distance from zero
Input: x = -8, y = 8
Output: True'''

x,y=-8,8
if (abs(x)-0)==(abs(y-0)):
    print("True")
else:
    print('False')
```

```python
'''6.Task 6: Check if two numbers are the same distance from zero
Input: x = -8, y = 8
Output: True'''

x,y=-8,8
print(abs(x)==abs(y))
    
```

```python
'''7.Task 7: Find the maximum absolute difference between two numbers in a list
Input: [3, -7, 4, -2]
Output: Maximum absolute difference = 11 (between -7 and 4)'''

a=[3, -7, 4, -2]
print(f"Maximun absolute difference = {(max(a)-min(a))} (between {max(a)} and {min(a)})")
```

```python
'''8.Task 8: Calculate absolute value of a complex number
Input: z = 3 + 4j
Output: 5.0'''

z=3+4j
print(abs(z))
```
