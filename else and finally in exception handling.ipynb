```python
#else
a=int(input("n1:"))
b=int(input("n2:"))
try:
    result=a/b
except ZeroDivisionError:
    print("cannot dividing by zero")
except ValueError:
    print("aplpha cannot dividing")
else:
    print("you answer is:",result)
    
```

```python
#finally.... it 1st perform the try block then perform the finally block in conclusion finally block is run if error 
#is occured or not occured it will bw run
a=int(input("n1:"))
b=int(input("n2:"))
try:
    result=a/b
except ZeroDivisionError:
    print("cannot dividing by zero")
except ValueError:
    print("aplpha cannot dividing")
finally:
    print("you answer is:",result)
    
```
